```python
from igraph import Graph, plot
import numpy as np
import matplotlib.pyplot as plt
```

```python
graph = Graph.Read_Adjacency("Tags_Coword_Network.txt")
```

Calculating centralities:

```python
degree_centrality = graph.degree()
closeness_centrality = graph.closeness()
pagerank_centrality = graph.pagerank()
```

Drawing Scatter Plots:

```python
plt.scatter(degree_centrality, closeness_centrality)
plt.title("Degree Centrality vs. Closeness Centrality")
plt.xlabel("Degree Centrality")
plt.ylabel("Closeness Centrality")
plt.show()
```

```python
plt.scatter(degree_centrality, pagerank_centrality)
plt.title("Degree Centrality vs. Pagerank Centrality")
plt.xlabel("Degree Centrality")
plt.ylabel("Pagerank Centrality")
plt.show()
```

```python
plt.scatter(closeness_centrality, pagerank_centrality)
plt.title("Closeness Centrality vs. Pagerank Centrality")
plt.xlabel("Closeness Centrality")
plt.ylabel("Pagerank Centrality")
plt.show()
```
